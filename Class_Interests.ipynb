```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Read in   data
class_data = pd.read_csv('Data/classinterests.txt', delimiter='\t')

# Year Distribution Plot
sns.countplot(x='Year', data=class_data)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Year Distribution of Class Interests')
plt.show()


# PRogram Distribution Plot
sns.countplot(x='Program', data=class_data)
plt.xlabel('Program')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.title('Program Distribution of Class Interests')
plt.show()
```
