### Standardizing Gene Expression Data

```python
import pandas as pd 
```

```python
matrix = pd.read_csv('Data/GSE5859_exprs.csv', index_col=0)

gene_rows = matrix.sub(matrix.mean(axis=1), axis=0)

gene_cols = gene_rows.sub(gene_rows.mean(axis=0), axis=1)

gene_std = gene_cols.div(gene_cols.std(axis=0), axis=1)

gene_std.head()
```
